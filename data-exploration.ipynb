```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
data
```

```python
data.columns
```

```python
data_test = pd.read_csv('test.csv')
data_test
```

```python
#convert data to dataframe
df_train = pd.DataFrame(data=data, columns=data.columns)
df_train.head()
```

```python
#looking for missing values
missing_values = data.isnull().sum()
missing_values
```

```python
data.shape
```

```python
# Replace missing values with "None" in 'agent' and 'company' columns
df_train['agent'].fillna("None", inplace=True)
df_train['company'].fillna("None", inplace=True)

# Now the missing values have been replaced with "None"
print(df_train['agent'].isnull().sum())  # Should output 0
print(df_train['company'].isnull().sum())  # Should output 0
```

```python
df_train.head()
```

```python
#select columns that will be used
selected_columns=['is_canceled','lead_time','arrival_date_month','stays_in_weekend_nights','stays_in_week_nights','adults','children','babies','previous_cancellations','previous_bookings_not_canceled','reserved_room_type','assigned_room_type','booking_changes','deposit_type','agent','company','customer_type','total_of_special_requests','adr','required_car_parking_spaces']
df_train = df_train[selected_columns]
df_train.head()
```

```python
#data cleaning by fill the missing values by mode
# Calculate the mode of the 'babies' column
babies_mode = df_train['babies'].mode()[0]

# Fill missing values in 'babies' with the mode
df_train['babies'].fillna(babies_mode, inplace=True)

# Now the missing values in 'babies' have been filled with the mode
print(df_train['babies'].isnull().sum()) 
```

```python
#data cleaning by fill the missing values by mode
# Calculate the mode of the 'babies' column
child = df_train['children'].mode()[0]

# Fill missing values in 'children' with the mode
df_train['children'].fillna(child, inplace=True)

# Now the missing values in 'children' have been filled with the mode
print(df_train['children'].isnull().sum()) 
```

```python
df_train.head()
```

```python
#feature engineering to create new feature cross
# Create the 'family_vs_solo_travel' feature
df_train['family_vs_solo_travel'] = df_train.apply(lambda row: 1 if row['children'] > 0 or row['babies'] > 0 else 0, axis=1)

# Now the 'family_vs_solo_travel' feature has been added to the DataFrame
print(df_train[['children', 'babies', 'family_vs_solo_travel']].head())
```

```python
#feature engineering to create new feature cross
# Calculate historical cancellation rate for 'agent'
agent_cancel_rates = df_train.groupby('agent')['previous_cancellations', 'previous_bookings_not_canceled'].sum()
agent_cancel_rates['historical_cancellation_rate'] = agent_cancel_rates['previous_cancellations'] / (agent_cancel_rates['previous_cancellations'] + agent_cancel_rates['previous_bookings_not_canceled'])
agent_cancel_rates = agent_cancel_rates.drop(['previous_cancellations', 'previous_bookings_not_canceled'], axis=1)

# Calculate historical cancellation rate for 'company'
company_cancel_rates = df_train.groupby('company')['previous_cancellations', 'previous_bookings_not_canceled'].sum()
company_cancel_rates['historical_cancellation_rate'] = company_cancel_rates['previous_cancellations'] / (company_cancel_rates['previous_cancellations'] + company_cancel_rates['previous_bookings_not_canceled'])
company_cancel_rates = company_cancel_rates.drop(['previous_cancellations', 'previous_bookings_not_canceled'], axis=1)

# Merge historical cancellation rates back into the main dataset
df_train = df_train.merge(agent_cancel_rates, left_on='agent', right_index=True, how='left')
df_train = df_train.merge(company_cancel_rates, left_on='company', right_index=True, how='left')
```

```python
df_train.drop('customer_history', axis=1)
```

```python
df_train['agent'].head(50)
```

```python
missing_values = df_train.isnull().sum()
missing_values
```

```python
df_train['historical_cancellation_rate_x'].fillna(0.0, inplace=True)
df_train['historical_cancellation_rate_y'].fillna(0.0, inplace=True)
```

```python
missing_values = df_train.isnull().sum()
missing_values
```

```python
#convert object dtypes to categorical
# List of columns to convert to categorical
categorical_columns = ['arrival_date_month', 'reserved_room_type', 'assigned_room_type', 'deposit_type', 'agent', 'company', 'customer_type']

# Convert columns to categorical data type
df_train[categorical_columns] = df_train[categorical_columns].astype('category')

# Check the updated data types
print(df_train.dtypes)
```

```python
#convert categorical dtypes to numeric
# Columns to convert to numeric
numeric_conversion_columns = ['historical_cancellation_rate_x', 'historical_cancellation_rate_y']

# Remove non-numeric characters and convert to numeric
for column in numeric_conversion_columns:
    df_train[column] = pd.to_numeric(df_train[column].str.replace('[^0-9.]', ''), errors='coerce')

# Check the updated data types
print(df_train.dtypes)
```

```python
#label encoder for the categorical dtypes
from sklearn.preprocessing import LabelEncoder

# Assuming your DataFrame is named 'data'
label_encoder = LabelEncoder()
# Convert categorical columns to numeric representation
for column in categorical_columns:
    df_train[column] = label_encoder.fit_transform(df_train[column].astype(str))
```

```python
columns = ['is_canceled','customer_history']
#separate the features and target/label
X = df_train.drop(columns=columns, axis=1)
y = df_train['is_canceled']
```

```python
missing_values = X.isnull().sum()
missing_values
```

```python
X['historical_cancellation_rate_x'].fillna(0.0, inplace=True)
X['historical_cancellation_rate_y'].fillna(0.0, inplace=True)
```

```python
#modelling using logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

logreg = LogisticRegression(max_iter=1000, solver='lbfgs')

cross_val_scores = cross_val_score(logreg, X, y, cv=5, scoring='f1')

# Print cross-validation scores
print("Cross-Validation F1 Scores:", cross_val_scores)
print("Mean F1 Score:", cross_val_scores.mean())
```

```python
#modelling using decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

DT = DecisionTreeClassifier()

cross_val_scores = cross_val_score(DT, X, y, cv=5, scoring='f1')

# Print cross-validation scores
print("Cross-Validation F1 Scores:", cross_val_scores)
print("Mean F1 Score:", cross_val_scores.mean())
```

```python
#modelling using random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

RF = RandomForestClassifier()

cross_val_scores = cross_val_score(RF, X, y, cv=5, scoring='f1')

# Print cross-validation scores
print("Cross-Validation F1 Scores:", cross_val_scores)
print("Mean F1 Score:", cross_val_scores.mean())
```
